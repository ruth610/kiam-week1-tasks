```python
news["date"] = pd.to_datetime(news["date"]).dt.date
stock.index = pd.to_datetime(stock.index).date
news["headline"] = news["headline"].astype(str)
```

```python
from textblob import TextBlob

def get_sentiment(text):
    return TextBlob(text).sentiment.polarity  # -1 to +1

news["sentiment"] = news["headline"].apply(get_sentiment)
```

```python
stock["returns"] = stock["Close"].pct_change()
```

```python
daily_sentiment = news.groupby("date")["sentiment"].mean()
```

```python
combined = pd.merge(
    daily_sentiment,
    stock["returns"],
    left_index=True,
    right_index=True,
    how="inner"
)
```

```python
correlation = combined["sentiment"].corr(combined["returns"], method="pearson")
correlation
```

```python
plt.scatter(combined["sentiment"], combined["returns"])
plt.xlabel("Sentiment")
plt.ylabel("Daily Returns")
plt.title("Sentiment vs Stock Returns")
plt.show()
```
