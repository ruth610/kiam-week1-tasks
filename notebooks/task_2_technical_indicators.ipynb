```python
import pandas as pd

df = pd.read_csv("raw_analyst_ratings.csv", parse_dates=["Date"])
df.set_index("Date", inplace=True)

df.head()
```

```python
import talib

df["SMA_20"] = talib.SMA(df["Close"], timeperiod=20)
df["EMA_20"] = talib.EMA(df["Close"], timeperiod=20)
```

```python
df["RSI_14"] = talib.RSI(df["Close"], timeperiod=14)
```

```python
df["MACD"], df["MACD_signal"], df["MACD_hist"] = talib.MACD(
    df["Close"], 
    fastperiod=12, 
    slowperiod=26, 
    signalperiod=9
)
```

```python
import pynance as pn

returns = pn.stats.returns(df["Close"])
vol = pn.stats.volatility(df["Close"])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,6))
plt.plot(df["Close"], label="Close")
plt.plot(df["SMA_20"], label="SMA 20")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14,4))
plt.plot(df["RSI_14"], label="RSI 14")
plt.axhline(70, linestyle="--")
plt.axhline(30, linestyle="--")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14,4))
plt.plot(df["MACD"], label="MACD")
plt.plot(df["MACD_signal"], label="Signal")
plt.bar(df.index, df["MACD_hist"])
plt.legend()
plt.show()
```
